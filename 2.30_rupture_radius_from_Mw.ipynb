_Stress drops 1Ïƒ usually vary between 1 and 10 MPa. Use the relation $$\Delta\sigma = \frac{7M_0}{16r^3}$$ between stress drop, scalar moment, and (circular) rupture radius to plot the expected radii for varying magnitudes for given stress drop. Carefully check physical units!_

See exercise 2.29 for relation between scalar moment and moment magnitude.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dsigma = 5E6                        # Pa = N / m^2
Mw = np.linspace(0, 10, 100)
M0 = 10**((3/2) * Mw + 9.1)         # J  = N * m
r = (7*M0 / (16 * dsigma))**(1/3)   # [ N*m / (N/m^2) ]^(1/3) = m

plt.plot(Mw, r, 'slateblue', label = r'$r = (\frac{7M_0}{16\Delta\sigma})^{1/3}$')
plt.yscale('log')
plt.legend(fontsize = 16)
plt.xlabel('Moment magnitude, $M_w$ (J)', fontsize = 14)
plt.ylabel('Rupture radius, $r$ (m)', fontsize = 14)
plt.title(r'Rupture radius at $\Delta \sigma$ = 5 MPa', fontsize = 16)
plt.grid()
```
