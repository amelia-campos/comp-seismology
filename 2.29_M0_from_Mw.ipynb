# Scalar moment from energy magnitude
_Computational Seismology, Igel (2016)_ <br>
Exercise 2.29

_Write a program that plots the scalar moment M0 as a function of energy
magnitude Mw._

$M_w = \frac{2}{3} (\text{log}_{10}M_0 - 9.1)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Mw = np.linspace(0, 10, 100)
M0 = 10**((3/2) * Mw + 9.1)

plt.plot(Mw, M0, 'b', label = r'$M_0 = 10^{\frac{3}{2} M_w + 9.1}$')
plt.yscale('log')
plt.xlabel('$M_w$', fontsize = 14)
plt.ylabel('log($M_0$) (J)', fontsize = 14)
plt.title('Scalar moment', fontsize = 16)
plt.legend(fontsize = 16)
plt.grid()
```
